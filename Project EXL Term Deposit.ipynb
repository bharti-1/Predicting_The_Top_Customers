## Removal of outlier

```python
print(data_1["balance"].quantile(0.5))
print(data_1["balance"].quantile(0.95))
print(data_1["balance"].quantile(0.90))
print(data_1["balance"].quantile(0.99))
print(data_1["balance"].quantile(0.75))
```

```python
percentile25 = data_1['balance'].quantile(0.25)
percentile75 = data_1['balance'].quantile(0.75)
iqr=percentile75-percentile25
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
data_1["balance"]=np.where(data_1["balance"]>upper_limit,upper_limit,np.where(data_1["balance"]<lower_limit,lower_limit,data_1["balance"]))
```

```python
sns.boxplot(y="balance",data=data_1)
```

```python
percentile25 = data_1['customer_age'].quantile(0.25)
percentile75 = data_1['customer_age'].quantile(0.75)
iqr=percentile75-percentile25
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
data_1["customer_age"]=np.where(data_1["customer_age"]>upper_limit,upper_limit,np.where(data_1["customer_age"]<lower_limit,lower_limit,data_1["customer_age"]))
```

```python
sns.boxplot(y="customer_age",data=data_1)
```

```python
percentile25 = data_1['num_contacts_in_campaign'].quantile(0.25)
percentile75 = data_1['num_contacts_in_campaign'].quantile(0.75)
iqr=percentile75-percentile25
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
data_1["num_contacts_in_campaign"]=np.where(data_1["num_contacts_in_campaign"]>upper_limit,upper_limit,np.where(data_1["num_contacts_in_campaign"]<lower_limit,lower_limit,data_1["num_contacts_in_campaign"]))
```

```python
sns.boxplot(y="num_contacts_in_campaign",data=data_1)
```

## Applying SMOTE to balance the unbalance data.
Synthetic Minority Oversampling Technique or SMOTE is another technique to oversample the minority class. Simply adding duplicate records of minority class often dont add any new information to the model. In SMOTE new instances are synthesized from the existing data. If we explain it in simple words, SMOTE looks into minority class instances and use k nearest neighbor to select a random nearest neighbor, and a synthetic instance is created randomly in feature space.

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(sampling_strategy='minority')
X_sm,y_sm = smote.fit_resample(X_scaler,Y)
```

```python
y_sm.shape
```

```python
X_sm.shape
```

```python
X_train_1, X_test, y_train_1, y_test = train_test_split(X_sm, y_sm, test_size=.3,random_state=15,stratify=y_sm)
```

### Find the confusion matrix
A Confusion matrix is an N x N matrix used for evaluating the performance of a classification model, where N is the number of target classes. The matrix compares the actual target values with those predicted by the machine learning model.



Binary Classification Problem (2x2 matrix)
A good model is one which has high TP and TN rates, while low FP and FN rates.
If you have an imbalanced dataset to work with, its always better to use confusion matrix as your evaluation criteria for your machine learning model.
